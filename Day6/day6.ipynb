```python
import pandas as pd
```

```python
weight = [37,48,42,39,41,38,40,40,100]
weight_table = pd.Series(weight,index=["Hoa","Nam","Hải","Linh","Nhân","Phương","Thu","Vũ","Bảo"])
weight_table
```

```python
print("Mean: ", weight_table.mean())
print("Median: ", weight_table.median())
print("Mode: ", weight_table.mode()[0])

print(weight_table[weight_table == weight_table.mode()[0]])
# print(weight_table = "Hoa")
# print(weight_table.nlargest(1))
```
