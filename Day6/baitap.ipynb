```python
# Buoi 1
import pandas as pd
data = pd.DataFrame({
    "gio_hoc":[5,8,10,12,15],
    "chuyen_can":[80,90,95,100,98],
    "hoc_them":[0,1,1,1,1],
    "diem_toan":[5.5, 7.0, 8.0, 9.0, 9.5]
})
data.to_csv("du_lieu.csv",index=False)
data = pd.read_csv("du_lieu.csv")
print("Du lieu HS:")
print(data)

tong = data["diem_toan"].sum()
hoc_sinh = len(data)
trung_binh = tong/hoc_sinh
print("Diem Toan trung binh:", trung_binh)
print("HS co diem toan > trung binh:")
hoc_sinh_gioi = data[data["diem_toan"] > trung_binh]
print(hoc_sinh_gioi)

print("NX:")
print("- HS co diem cao thuong hoc nhieu gio hon")
print("- HS co chuyen can cao thuong dat KQ tot hon")
print("- Phan lon HS diem cao deu co hoc them")
```

```python
# Buoi 2:
# Yeu cau 1:
numbers = list(range(1, 101))
print(numbers)
```

```python
# Buoi 2:
# Yeu cau 2:
def lay_so_le(ds):
    ket_qua = []
    for so in ds:
        if so % 2 != 0:
            ket_qua.append(so)
    return ket_qua
so_le = lay_so_le(numbers)
print(so_le)
```
